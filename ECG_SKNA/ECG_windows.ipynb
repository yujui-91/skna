```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
def process_folder(folder_path, output_csv_name):
    """
    讀取指定資料夾內的 .npy 檔，統計 ID 與視窗數，並存成 CSV。
    """
    # 檢查資料夾是否存在
    if not os.path.exists(folder_path):
        print(f"❌ 找不到資料夾: {folder_path}")
        return

    files = [f for f in os.listdir(folder_path) if f.endswith('.npy')]
    results = []

    print(f"📂 正在處理資料夾: {os.path.basename(folder_path)} (共 {len(files)} 個檔案)...")

    for filename in tqdm(files):
        file_path = os.path.join(folder_path, filename)
        
        try:
            # 只讀取 header 資訊，速度快
            data = np.load(file_path, mmap_mode='r')
            
            # 抓 ID
            p_id = filename.replace('.npy', '')
            
            # 抓視窗數 (第一維度)
            win_count = data.shape[0]
            
            results.append({'ID': p_id, 'Windows': win_count})
            
        except Exception as e:
            print(f"⚠️ 讀取錯誤 {filename}: {e}")

    # 轉成 DataFrame
    df = pd.DataFrame(results)
    
    save_path = os.path.join(os.path.dirname(folder_path), output_csv_name)
    df.to_csv(save_path, index=False)
    
    print(f"✅ {output_csv_name} 已儲存 (筆數: {len(df)})")
    print(df.head()) # 顯示前幾筆看一下
    print("-" * 30)
```

```python
base_path = r"D:\M143020071\raw_data_result\SKNA_signal\ch1\sr10000_500_2000_15s_ECG_signal_rpeak_5_10min_longer_1000pts_mr_9win"

# 2. 設定兩個子資料夾路徑
mace_path = os.path.join(base_path, "mace_zscore")
non_mace_path = os.path.join(base_path, "non_mace_zscore")

# 3. 分別執行 (分開 load，分開存)
process_folder(mace_path, "mace_windows_info.csv")
process_folder(non_mace_path, "non_mace_windows_info.csv")
```
