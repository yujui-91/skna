```python
import numpy as np
import os
from tqdm import tqdm
```

```python
non_mace_dir = r'D:\M143020071\raw_data_result\SKNA_signal\ch1\sr10000_500_2000_15s_ECG_signal_rpeak_5_10min_longer_v1_mr\non_mace_zscore/'
mace_dir = r'D:\M143020071\raw_data_result\SKNA_signal\ch1\sr10000_500_2000_15s_ECG_signal_rpeak_5_10min_longer_v1_mr\mace_zscore/'
save_non_mace_dir = r'D:\M143020071\raw_data_result\SKNA_signal\ch1\sr10000_500_2000_15s_ECG_signal_rpeak_5_10min_longer_1000pts_mr_9win\non_mace_zscore/'
save_mace_dir = r'D:\M143020071\raw_data_result\SKNA_signal\ch1\sr10000_500_2000_15s_ECG_signal_rpeak_5_10min_longer_1000pts_mr_9win\mace_zscore/'
```

```python
for save_path in [save_non_mace_dir, save_mace_dir]:
    if not os.path.exists(save_path):
        os.makedirs(save_path)
        print(f'Created directory: {save_path}')
    else:
        print(f'Directory already exists: {save_path}')
```

```python
for data_dir, save_dir in zip([non_mace_dir, mace_dir], [save_non_mace_dir, save_mace_dir]):
    for filename in tqdm(os.listdir(data_dir),desc=f'Processing files in {os.path.basename(os.path.normpath(data_dir))}'):
        if filename.endswith('.npy'):
            try:
                raw_data = np.load(os.path.join(data_dir, filename))
                labels = raw_data[:, 0].astype(np.int32)
                signal = raw_data[:, 1:].astype(np.float32)
                signal_downsampled = signal.reshape(signal.shape[0], -1, 1000)
                mav_signal = np.mean(np.abs(signal_downsampled), axis=2)
                mav_signal = np.hstack((labels.reshape(-1, 1), mav_signal)).astype(np.float32)#變成了一個 20x1 的 2D 矩陣 (直欄向量) 為了hstack 維度要都是2維
                np.save(os.path.join(save_dir, filename), mav_signal[:9,:]) #取前n個特徵
            except Exception as e:
                print(f'Error processing file {filename}: {e}')
    print(f'Saved processed file: {save_dir}')
            
```
