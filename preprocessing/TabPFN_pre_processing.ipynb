```python
import pandas as pd
```

```python
pre_process_path = r'V:\dunwei\MACE\dataset\pre_process/'
save_data_path = r'V:\dunwei\MACE\dataset\pre_process\five_feature/'
imputed_data_path = r'V:\dunwei\MACE\dataset\reversed/'
```

```python
feature = ['count_v_ar', 'count_copd', 'smoke_try', 'count_dm', 'male']
research_id = ['Transformation_ID']
label = ['mace']
```

```python
mode = 1
"""
mode = 1 ---> raw data pre-processing
mode = 2 ---> imputed data pre-processing
"""
```

```python
raw_data = pd.read_csv(pre_process_path + 'MACE_raw_data.csv')
display(raw_data)
```

```python
if mode == 1:
    print("raw data pre-processing")
    raw_data = pd.read_csv(pre_process_path + "MACE_research_id_transformation.csv")
    raw_data = raw_data[research_id + feature + label]
    display(raw_data)

    H = []
    P = []
    for idx, row in raw_data.iterrows():
        if row['mace'] == 0:
            H.append(row)
        else:
            P.append(row)

    H_df = pd.DataFrame(H).reset_index(drop=True)
    P_df = pd.DataFrame(P).reset_index(drop=True)

    H_df.to_csv(save_data_path + "H.csv", index=False)
    P_df.to_csv(save_data_path + "P.csv", index=False)

elif mode == 2:
    print("imputed data pre-processing")
    imputed_data = pd.read_csv(imputed_data_path + "MACE_imputed_data.csv")
    # 依照編號由小到大排列
    imputed_data['SortKey'] = imputed_data.iloc[:, 0].str.extract(r'(\d+)').astype(int)  # 提取數字部分，轉成整數型別
    imputed_data_sorted = imputed_data.sort_values(by='SortKey')  # 根據數字排序
    imputed_data_sorted = imputed_data_sorted.drop(columns=['SortKey'])  # 刪除輔助排序欄位（可選）
    imputed_data_sorted = imputed_data_sorted.reset_index(drop=True)  # 重設索引

    H = []
    P = []
    for idx, row in imputed_data_sorted.iterrows():
        if row['mace'] == 0:
            H.append(row)
        else:
            P.append(row)

    H_df = pd.DataFrame(H).reset_index(drop=True)
    P_df = pd.DataFrame(P).reset_index(drop=True)

    H_df.to_csv(save_data_path + "H_imputed.csv", index=False)
    P_df.to_csv(save_data_path + "P_imputed.csv", index=False)

else:
    raise ValueError("error mode")

print(f"save_path: {save_data_path}")
```
