```python
import pandas as pd
import os
import glob
```

```python
raw_data_path = r"F:\M143020071\raw_data/"
id_path = r"D:\M143020071\raw_data_result\tabular_data/"
save_data_path = r"D:\M143020071\raw_data_result/"
```

```python
raw_data = pd.read_csv(os.path.join(raw_data_path, "MACE 資料登錄(新增VT)_killip finial.csv"))
id_corr = pd.read_csv(os.path.join(id_path, "MACE_id_correlation.csv"))
display(id_corr)



print("success")
```

```python
h = []
p = []
id_num = '研究編號'
label = 'mace'

for idx, row in raw_data.iterrows():
    if  row[label] == 0:
        h.append(row[id_num])
    elif row[label] == 1:
        p.append(row[id_num])
    else:
        print(f"Error label in row {idx + 1}: {row[label]}, id: {row[id_num]}")

display(p)
```

```python
h_df = pd.DataFrame(h, columns=['research_ID'])
p_df = pd.DataFrame(p, columns=['research_ID'])
display(h_df)
```

```python
h_conversion = []
p_conversion = []
for idx, row in h_df.iterrows():
    for idx2, row2 in id_corr.iterrows():
        if row['research_ID'] == row2['Original_ID']:
            h_conversion.append(row2['Transformation_ID'])
            break
        else:
            pass

for idx, row in p_df.iterrows():
    for idx2, row2 in id_corr.iterrows():
        if row['research_ID'] == row2['Original_ID']:
            p_conversion.append(row2['Transformation_ID'])
            break
        else:
            pass
```

```python
h_conversion_df = pd.DataFrame(h_conversion, columns=['conversion_ID'])
p_conversion_df = pd.DataFrame(p_conversion, columns=['conversion_ID'])

h_corr_df = pd.concat([h_df, h_conversion_df], axis=1)
p_corr_df = pd.concat([p_df, p_conversion_df], axis=1)
h_corr_df.to_csv(save_data_path + "MACE_h_id.csv", index=False)
p_corr_df.to_csv(save_data_path + "MACE_p_id.csv", index=False)
display(h_corr_df.to_csv)
print(f'file saved in : {save_data_path}')
```
