```python
import pandas as pd
import os
```

```python
MACE_raw_data_dir = r'V:\dunwei\MACE\dataset\tabular_data/'
save_dir = r'V:\dunwei\MACE\dataset\tabular_data\static_feature/'
```

```python
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    print(f"Created directory: {save_dir}")
else:
    print(f"Directory already exists: {save_dir}")
```

```python
research_id = ['Transformation_ID']
feature = ['count_v_ar', 'count_copd', 'smoke_try', 'count_dm', 'male']
label = ['mace']
```

```python
MACE_raw_data = pd.read_csv(os.path.join(MACE_raw_data_dir, 'MACE_research_id_transformation.csv'))
display(MACE_raw_data)
```

```python
MACE_extract_raw_data = MACE_raw_data.loc[:, research_id + feature + label]
display(MACE_extract_raw_data)
MACE_extract_raw_data.to_csv(os.path.join(save_dir, 'MACE_preprocessed_tabular_data_5features.csv'), index=False)
```
