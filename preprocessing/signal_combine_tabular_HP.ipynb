```python
import pandas as pd
```

```python
file_path = r"V:\dunwei\MACE\dataset\pre_process\combine_feature/"
save_data_path = r"V:\dunwei\MACE\dataset\pre_process\combine_feature/"
```

```python
mode = 1
"""
1: Missing value
2: Non-missing value
"""
if mode == 1:
    data = pd.read_csv(file_path + "combine_tabular_data_missing_value.csv")
elif mode == 2:
    data = pd.read_csv(file_path + "combine_tabular_data_non_missing_value.csv")
display(data)
```

```python
header = data.columns.tolist()
data_arr = data.values
h = []
p = []
```

```python
for i in range(data_arr.shape[0]):
    if data_arr[i, -1] == 0:
        h.append(data_arr[i, :])
    elif data_arr[i, -1] == 1:
        p.append(data_arr[i, :])

h_df = pd.DataFrame(h, columns=header)
p_df = pd.DataFrame(p, columns=header)

if mode == 1:
    h_df.to_csv(save_data_path + "missing_value/combine_h_missing_value.csv", index=False)
    p_df.to_csv(save_data_path + "missing_value/combine_p_missing_value.csv", index=False)
elif mode == 2:
    h_df.to_csv(save_data_path + "non_missing_value/combine_h_non_missing.csv", index=False)
    p_df.to_csv(save_data_path + "non_missing_value/combine_p_non_missing.csv", index=False)

print(f"save path : {save_data_path}")
```
