```python
import pandas as pd
import os
import numpy as np
```

```python
file_path = r'G:\高醫心肌梗塞_dataset_20250725\OneDrive_1_2025-7-25'
save_path = r'V:\dunwei\MACE\dataset'
```

```python
raw_data = pd.read_csv(os.path.join(file_path, 'MACE 資料登錄(新增VT)_killip finial.csv'))
```

```python
H_raw_data = []
P_raw_data = []
for i in range(len(raw_data)):
    if raw_data.iloc[i]['mace'] == 0:
        H_raw_data.append(raw_data.iloc[i, :])
    elif raw_data.iloc[i]['mace'] == 1:
        P_raw_data.append(raw_data.iloc[i, :])

H_data_df = pd.DataFrame(H_raw_data)
P_data_df = pd.DataFrame(P_raw_data)

display(H_data_df)
display(P_data_df)


H_data_df.to_excel(os.path.join(save_path, 'H_raw_data.xlsx'), index=False)
P_data_df.to_excel(os.path.join(save_path, 'P_raw_data.xlsx'), index=False)
```
