```python
import numpy as np
import os
```

```python
data_dir = r"D:\M143020071\raw_data_result\SKNA_signal\ch1\sr10000_500_1000_20s_ECG_signal_rpeak_5_10min_longer_100pts_mr_20win\x_transform"
```

```python
X_all = np.load(os.path.join(data_dir, 'x_transform_all.npy'))
# 載入標籤 (N, ) -> 0是健康, 1是病人
y_all = np.load(os.path.join(data_dir, 'y_transform_all.npy'))

print(f"Loaded X_all shape: {X_all.shape}")
print(f"Loaded y_all shape: {y_all.shape}")
```

```python
non_mace_features = X_all[y_all == 0]
mace_features = X_all[y_all == 1]

print("Splitting complete:")
print(f"  - Non-MACE (Healthy) shape: {non_mace_features.shape}")
print(f"  - MACE (Patient) shape:     {mace_features.shape}")
```

```python
# 4. 存檔 (生成學長後續程式需要的檔案)
# ==========================================
save_path_h = os.path.join(data_dir, "non_mace.npy")
save_path_p = os.path.join(data_dir, "mace.npy")

np.save(save_path_h, non_mace_features.astype(np.float32))
np.save(save_path_p, mace_features.astype(np.float32))

print("========================================")
print(f"Saved successfully!")
print(f"  -> {save_path_h}")
print(f"  -> {save_path_p}")
```
