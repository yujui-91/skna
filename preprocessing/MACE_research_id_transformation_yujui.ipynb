```python
import pandas as pd
import os
```

```python
raw_data_path = r"F:\M143020071\raw_data/"
save_data_path = r"D:\M143020071\raw_data_result\tabular_data/"
raw_data = pd.read_csv(os.path.join(raw_data_path, "MACE 資料登錄(新增VT)_killip finial.csv"))
display(raw_data)
print("success")
```

```python
non_research_id = raw_data.iloc[:, 1:]
display(non_research_id)
print("success")
```

```python
new_id_list = []
for i in range(raw_data.shape[0]):
    id_number = f"MACE{i+1:03d}" # 格式化ID為MACE001, MACE002, ..., MACE999
    new_id_list.append(id_number)
```

```python
new_id_df = pd.DataFrame(new_id_list, columns=['Transformation_ID'])
raw_id_df = pd.DataFrame(raw_data.iloc[:, 0].values, columns=['Original_ID'])
corr_id_df = pd.concat([raw_id_df, new_id_df], axis=1)
corr_id_df.to_csv(save_data_path + "MACE_id_correlation.csv", index=False)
display(corr_id_df)
print("success")
```

```python
'''
這邊要再修改,因為Killip是羅馬數字,要轉成哪種型態,和VT的轉換
'''
new_data = pd.concat([new_id_df, non_research_id], axis=1)
new_data[['age']] = new_data[['age']].astype(int) # Raw data age有小數
display(new_data)
new_data.to_csv(save_data_path + "MACE_research_id_transformation.csv", index=False)
print('==================== success')
```
