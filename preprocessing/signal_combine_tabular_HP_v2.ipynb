```python
import pandas as pd
import os
```

```python
file_path = r"V:\dunwei\MACE\dataset\tabular_data\static_dynamic_data/"
save_data_path = r"V:\dunwei\MACE\dataset\tabular_data\static_dynamic_data/"
```

```python
if not os.path.exists(save_data_path):
    os.makedirs(save_data_path)
    print(f"Created directory: {save_data_path}")
else:
    print(f"{save_data_path} already exists.")
```

```python
mode = 1
"""
1: Missing value
2: Non-missing value (non use)
"""
if mode == 1:
    data = pd.read_csv(file_path + "MACE_static_dynamic_founder_askna_mr_missing_data.csv")
elif mode == 2:
    data = pd.read_csv(file_path + "combine_tabular_data_non_missing_value.csv")
    print('No create file')
display(data)
```

```python
header = data.columns.tolist()
data_arr = data.values
display(data_arr)
```

```python
h = []
p = []
for i in range(data_arr.shape[0]):
    if data_arr[i, -1] == 0:
        h.append(data_arr[i, :])
    elif data_arr[i, -1] == 1:
        p.append(data_arr[i, :])

h_df = pd.DataFrame(h, columns=header)
p_df = pd.DataFrame(p, columns=header)
display(h_df)
display(p_df)

if mode == 1:
    h_df.to_csv(save_data_path + "missing_data/static_dynamic_founder_askna_mr_h_missing_value.csv", index=False)
    p_df.to_csv(save_data_path + "missing_data/static_dynamic_founder_askna_mr_p_missing_value.csv", index=False)
elif mode == 2:
    h_df.to_csv(save_data_path + "non_missing_data/static_dynamic_founder_h_non_missing_value.csv", index=False)
    p_df.to_csv(save_data_path + "non_missing_data/static_dynamic_founder_p_non_missing_value.csv", index=False)

print(f"save path : {save_data_path}")
```
