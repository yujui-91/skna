```python
data_dir = r"D:\M143020071\raw_data_result\SKNA_signal\ch1\sr10000_500_1000_20s_ECG_signal_rpeak_5_10min_longer_v1_mr\x_transform/"
per_window_dir = r'D:\M143020071\raw_data_result\SKNA_signal\ch1\sr10000_500_1000_20s_ECG_signal_rpeak_5_10min_longer_v1_mr/'
save_data_path = r"D:\M143020071\raw_data_result\SKNA_signal\ch1\sr10000_500_1000_20s_ECG_signal_rpeak_5_10min_longer_v1_mr\x_transform/"
```

```python
train_per_window_data = pd.read_csv(per_window_dir + 'train_window_per_person.csv', header=0)
test_per_window_data = pd.read_csv(per_window_dir + 'test_window_per_person.csv', header=0)
train_non_mace_filename = pd.read_csv(per_window_dir + 'train_h_file_name.csv', header=0)
test_non_mace_filename = pd.read_csv(per_window_dir + 'test_h_file_name.csv', header=0)
train_mace_filename = pd.read_csv(per_window_dir + 'train_p_file_name.csv', header=0)
test_mace_filename = pd.read_csv(per_window_dir + 'test_p_file_name.csv', header=0)
display(train_per_window_data)
display(test_per_window_data)
display(train_non_mace_filename)
display(test_non_mace_filename)
display(train_mace_filename)
display(test_mace_filename)
```

```python
train_non_mace_ids_list = train_non_mace_filename['file_name'].values
test_non_mace_ids_list = test_non_mace_filename['file_name'].values
train_non_mace_masks = train_per_window_data['file_name'].isin(train_non_mace_ids_list)
test_non_mace_masks = test_per_window_data['file_name'].isin(test_non_mace_ids_list)
train_split = np.sum(train_per_window_data['window_per_person'][train_non_mace_masks].values, dtype=int)
test_split = np.sum(test_per_window_data['window_per_person'][test_non_mace_masks].values, dtype=int)
print(train_split, test_split)
```

```python
train = np.load(data_dir + "x_transform_train.npy")
test = np.load(data_dir + "x_transform_test.npy")
print(train.shape, test.shape)

non_mace = np.vstack([train[:train_split, :], test[:test_split, :]])
mace = np.vstack([train[train_split:, :], test[test_split:, :]])
display(non_mace.shape, mace.shape)
np.save(save_data_path + "non_mace.npy", non_mace.astype(np.float32))
np.save(save_data_path + "mace.npy", mace.astype(np.float32))

print(f"save path : {save_data_path}")
```
