```python
import pandas as pd
import os
import shutil
```

```python
def clean_and_sort(file_list):
    # 去除 .npy 副檔名
    cleaned = [os.path.splitext(f)[0] for f in file_list]
    
    # 根據數字排序（假設格式為 MACE001、MACE002...）
    sorted_list = sorted(cleaned, key=lambda x: int(x.replace("MACE", "")))

    return sorted_list
```

```python
file_path = r"D:\M143020071\raw_data_result\SKNA_signal\ch1\sr10000_500_2000_15s_ECG_signal_rpeak_5_10min_longer_1000pts_mr_9win/"
save_data_path = r"D:\M143020071\raw_data_result\SKNA_signal\ch1\sr10000_500_2000_15s_ECG_signal_rpeak_5_10min_longer_1000pts_mr_9win/"

train_ratio = 0.8

file_h = os.listdir(file_path + "non_mace_zscore/")
file_p = os.listdir(file_path + "mace_zscore/")

file_h = [f for f in file_h if f.endswith(".npy")]
file_p = [f for f in file_p if f.endswith(".npy")]

file_h = clean_and_sort(file_h)
file_p = clean_and_sort(file_p)

h_df = pd.DataFrame(file_h, columns=["file_name"])
p_df = pd.DataFrame(file_p, columns=["file_name"])

train_h = file_h[:int(len(file_h) * train_ratio)]
train_p = file_p[:int(len(file_p) * train_ratio)]
test_h = file_h[int(len(file_h) * train_ratio):]
test_p = file_p[int(len(file_p) * train_ratio):]

train_h_df  = pd.DataFrame(train_h, columns=["file_name"])
train_p_df  = pd.DataFrame(train_p, columns=["file_name"])
test_h_df  = pd.DataFrame(test_h, columns=["file_name"])
test_p_df  = pd.DataFrame(test_p, columns=["file_name"])

h_df.to_csv(save_data_path + "h_file_name.csv", index=False)
p_df.to_csv(save_data_path + "p_file_name.csv", index=False)
train_h_df.to_csv(save_data_path + "train_h_file_name.csv", index=False)
train_p_df.to_csv(save_data_path + "train_p_file_name.csv", index=False)
test_h_df.to_csv(save_data_path + "test_h_file_name.csv", index=False)
test_p_df.to_csv(save_data_path + "test_p_file_name.csv", index=False)
print("success")
```
