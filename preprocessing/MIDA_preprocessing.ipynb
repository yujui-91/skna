```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import joblib
```

```python
read_data_path = r"V:\dunwei\MACE\dataset\pre_process/"
save_data_path = r"V:\dunwei\MACE\dataset\pre_process/"
save_scaler_path = r"V:\dunwei\MACE\program\preprocessing\scaler/"
```

```python
feature = ['acei_use', 'arb_use', 'c1_use', 'c3_use', 'alpha_use', 'coag_use', 'plt_use', 'NSTEMI', 'STEMI',
           'age', 'male', 'bmi', 'count_dys', 'count_dm','count_htn', 'count_a_ar', 'count_v_ar', 'count_copd',
           'count_asthma','cancer', 'smoke_try', 'Alcohol_try', 'BetelNut_try', 'bb_prior','bb_during', 'bb_after',
           'statin_use','SBP_1', 'DBP_1', 'HR_1','count_HF', 'NA1b', 'NA2b', 'NA1s', 'NA2s',
           'NA1r', 'NA2r', 'TIMI']

numerical_feature = ['age', 'bmi', 'SBP_1','DBP_1', 'HR_1', 'NA1b', 'NA2b',
                    'NA1s', 'NA2s', 'NA1r', 'NA2r', 'TIMI']

category_feature = ['acei_use', 'arb_use', 'c1_use', 'c3_use', 'alpha_use', 'coag_use', 'plt_use', 'NSTEMI', 'STEMI',
                    'count_dys', 'count_dm', 'count_htn', 'count_a_ar', 'count_v_ar', 'count_copd', 'count_asthma',
                    'cancer', 'bb_prior', 'bb_during','bb_after', 'statin_use', 'count_HF']

encoding_feature = ['male', 'smoke_try', 'Alcohol_try', 'BetelNut_try']

research_id = ['Transformation_ID']

label = ['mace']
```

```python
raw_data = pd.read_csv(read_data_path + "MACE_research_id_transformation.csv")
display(raw_data)
```

```python
raw_data = raw_data[research_id + feature + label]
raw_data.to_csv(save_data_path + "MACE_raw_data.csv", index=False)
display(raw_data)
```

```python
train_df = raw_data[~raw_data.isna().any(axis=1)].reset_index(drop=True) # 留下沒有缺失值得當作train
display(train_df)
test_df = raw_data[raw_data.isna().any(axis=1)].reset_index(drop=True)   # 留下有缺失值得當作test
display(test_df)
mask = test_df.isna().astype(int) # 可以將missing value補為1，記錄缺失值位置
display(mask)
train_df.to_csv(save_data_path + "MACE_raw_train.csv", index=False)
test_df.to_csv(save_data_path + "MACE_raw_test.csv", index=False)
```

```python
re_order_raw_data = pd.concat([train_df, test_df], axis=0).reset_index(drop=True) # Test data 加在 Train data 後面
display(re_order_raw_data)
re_order_raw_data.to_csv(save_data_path + "MACE_re_order_raw_data.csv", index=False)
```

```python
numerical_features = re_order_raw_data[numerical_feature]
display(numerical_features)
categorical_features = re_order_raw_data[category_feature]
display(categorical_features)
encoding_features = re_order_raw_data[encoding_feature]
display(encoding_features)
```

```python
numerical_placeholder = numerical_features.fillna(numerical_features.mean()) # 使用平均值填補數值型特徵的缺失值
display(numerical_placeholder)
categorical_placeholder = categorical_features.apply(lambda x: x.fillna(x.mode()[0]), axis = 0) # 使用眾數填補類別型特徵的缺失值
display(categorical_placeholder)
encoding_placeholder = encoding_features.apply(lambda x: x.fillna(x.mode()[0]), axis = 0) # 使用眾數填補編碼型特徵的缺失值
display(encoding_placeholder)
```

```python
scaler = MinMaxScaler()
scaler = scaler.fit(numerical_placeholder) # 生成min(x)和max(x)，做 Normalization
numerical_std = scaler.transform(numerical_placeholder)
numerical_std_df = pd.DataFrame(numerical_std, columns=numerical_feature)
display(numerical_std_df)
save_scaler = joblib.dump(scaler, save_scaler_path + "MACE_scaler.joblib") # 再轉回去
print(f'MACE_scaler saved to {save_scaler[0]}')
```

```python
encoded_features = pd.get_dummies(encoding_placeholder, columns=encoding_feature, dtype=int) # get_dummies轉成one-hot encoding method
display(encoded_features)
```

```python
std_encoding_data = pd.concat([re_order_raw_data[research_id], numerical_std_df, categorical_placeholder,
                               encoded_features, re_order_raw_data[label]], axis=1)
display(std_encoding_data)
```

```python
train_num = train_df.shape[0]
print(train_num)
train_std_encoding = std_encoding_data.iloc[:train_num, :]
display(train_std_encoding)
test_std_encoding = std_encoding_data.iloc[train_num:, :]
display(test_std_encoding)
```

```python
std_encoding_data.to_csv(save_data_path + "MACE_std_encoding.csv", index=False)
print(f'std_encoding_data saved to {save_data_path  + "MACE_std_encoding.csv"}')
train_std_encoding.to_csv(save_data_path + "MACE_std_encoding_train.csv", index=False)
print(f'train_std_encoding saved to {save_data_path + "MACE_std_encoding_train.csv"}')
test_std_encoding.to_csv(save_data_path + "MACE_std_encoding_test.csv", index=False)
print(f'test_std_encoding saved to {save_data_path + "MACE_std_encoding_test.csv"}')
mask.to_csv(save_data_path + "MACE_mask.csv", index=False)
print(f'mask saved to {save_data_path + "MACE_mask.csv"}')
```
